Producer -> Channel IN -> Worker pool -> Chnnel OUT -> Consumer

```go
import (
    "fmt"
    "math/rand"
    "time"
)
```

```go
func echoWorker(in, out chan int) {
    for {
        n := <- in // waiting for the queue
        
        time.Sleep(time.Duration(rand.Intn(3000)) * time.Millisecond)
        
        out <- n // send it out
    }
}

func produce(ch chan <- int) {
    i := 0
    for {
        fmt.Printf("-> Send job: %d\n", i)
        ch <- i
        i++
    }
}
```

```go
in := make(chan int)
out := make(chan int)

go echoWorker(in, out)
go produce(in)

for n := range out { // consumer
    fmt.Printf("<- Rec job: %d\n", n)
}
```

```go
// Create multiple workers

in := make(chan int)
out := make(chan int)

for i := 0; i < 2; i++ {
    go echoWorker(in, out)
}

go produce(in)

for n := range out { // consumer
    fmt.Printf("<- Rec job: %d\n", n)
}
```

```go
// Create multiple workers

in := make(chan int)
out := make(chan int)

for i := 0; i < 100; i++ { // Increase performance!
    go echoWorker(in, out)
}

go produce(in)

for n := range out { // consumer
    fmt.Printf("<- Rec job: %d\n", n)
}
```
